## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/netzzwergeGithub/bash_buddy.git
```

or from [conda][conda]

```sh
$ conda install -c netzzwergeGithub bash_buddy
```

or from [pypi][pypi]


```sh
$ pip install bash_buddy
```
