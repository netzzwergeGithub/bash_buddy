```
#| export
def tmux_history():
    o = shell('tmux capture-pane -p', capture_output=True, text=True)
    return o.stdout or o.stderr
    
```

```
print(tmux_history())
```

```
#| export
def fmt_query(query):
    q = f"""
<tmux_history>
{tmux_history()}
</tmux_history>
<user_query>
{query}
</user_query>"""
    return q
```

```
query = fmt_query('How could I change my branch?')
print(query)
```
